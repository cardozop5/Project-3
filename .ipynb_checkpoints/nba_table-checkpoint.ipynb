```python
import pandas as pd
from splinter import Browser
import requests

url='https://stats.nba.com/teams/traditional/?sort=W_PCT&dir=-1&Season=2018-19&SeasonType=Regular%20Season'

website_url=requests.get('https://stats.nba.com/teams/traditional/?sort=W_PCT&dir=-1&Season=2018-19&SeasonType=Regular%20Season').text

from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python
browser = Browser('chrome', headless=False)
```

```python
browser.visit(url)

tables = pd.read_html(browser.html)
tables
```
